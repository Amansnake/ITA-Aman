```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming df_metrics is already populated from previous steps
# If not, the following lines would recreate it based on global and cluster metrics
metrics_data = []

# Global Model Metrics (using the values from the executed cell 94761467 outputs)
metrics_data.append({
    'Model': 'RF Global',
    'Type': 'Global',
    'RMSE': 0.0211,
    'MAE': 0.0145,
    'R-squared': 0.9994
})
metrics_data.append({
    'Model': 'GBR Global',
    'Type': 'Global',
    'RMSE': 0.0209,
    'MAE': 0.0145,
    'R-squared': 0.9994
})

# Cluster-Based Model Metrics (using the values from the executed cell eeb6e1f4 outputs)
rf_cluster_metrics_hardcoded = {
    0: {'RMSE': 0.0540, 'MAE': 0.0222, 'R-squared': 0.9836},
    1: {'RMSE': 0.0283, 'MAE': 0.0195, 'R-squared': 0.9995},
    2: {'RMSE': 0.0249, 'MAE': 0.0178, 'R-squared': 0.9992}
}
gbr_cluster_metrics_hardcoded = {
    0: {'RMSE': 0.0453, 'MAE': 0.0184, 'R-squared': 0.9885},
    1: {'RMSE': 0.0229, 'MAE': 0.0172, 'R-squared': 0.9997},
    2: {'RMSE': 0.0240, 'MAE': 0.0175, 'R-squared': 0.9993}
}

for cluster_id, metrics in rf_cluster_metrics_hardcoded.items():
    metrics_data.append({
        'Model': f'RF Cluster {cluster_id}',
        'Type': f'Cluster {cluster_id}',
        'RMSE': metrics['RMSE'],
        'MAE': metrics['MAE'],
        'R-squared': metrics['R-squared']
    })

for cluster_id, metrics in gbr_cluster_metrics_hardcoded.items():
    metrics_data.append({
        'Model': f'GBR Cluster {cluster_id}',
        'Type': f'Cluster {cluster_id}',
        'RMSE': metrics['RMSE'],
        'MAE': metrics['MAE'],
        'R-squared': metrics['R-squared']
    })

df_metrics = pd.DataFrame(metrics_data)

# Sort models for better visualization, global models first, then by cluster
df_metrics['Sort_Key'] = df_metrics['Model'].apply(lambda x: 0 if 'Global' in x else int(x.split()[-1]))
df_metrics = df_metrics.sort_values(by='Sort_Key').drop(columns='Sort_Key')

print('Consolidated Metrics DataFrame:')
display(df_metrics)

# Plotting function for metrics (copied from cell 67c8977d to regenerate the plots)
def plot_metrics_bar(df, metric_col, title, ylabel):
    plt.figure(figsize=(12, 6))
    bars = plt.bar(df['Model'], df[metric_col], color=['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan'])
    plt.title(title)
    plt.xlabel('Model')
    plt.ylabel(ylabel)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add metric values on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.001, round(yval, 4), ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

# Generate plots for each metric
plot_metrics_bar(df_metrics, 'RMSE', 'Model RMSE Comparison', 'RMSE')
plot_metrics_bar(df_metrics, 'MAE', 'Model MAE Comparison', 'MAE')
plot_metrics_bar(df_metrics, 'R-squared', 'Model R-squared Comparison', 'R-squared')
```
